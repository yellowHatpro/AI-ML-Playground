```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 14
plt.rcParams['font.family'] = 'bold'
plt.rcParams['axes.labelweight'] = 'bold'
plt.style.use('seaborn-whitegrid')
```

```python
cars_df = pd.read_csv('./train.csv')
```

```python
cars_df.head()
```

```python
cars_first_df = cars_df[["CarName","doornumber","carbody","wheelbase","carlength","carwidth","carheight","curbweight","cylindernumber","enginesize","fuelsystem","boreratio","stroke","compressionratio","horsepower","peakrpm","citympg","highwaympg","price"]]
cars_first_df.head()
```

```python
def printXYCharts(df, x, y):
    sns.lmplot(x=x, y=y, data=df, aspect=2)
    plt.xlabel(x)
    plt.ylabel(y)
    plt.title(f'{x} vs {y}')
    plt.show()
```

```python
list_of_attributes = ["wheelbase","carlength","carwidth","carheight","curbweight","enginesize","boreratio","stroke","compressionratio","horsepower","peakrpm","citympg","highwaympg","price"]
for attribute in list_of_attributes:
    printXYCharts(cars_first_df, attribute, "price")
```
