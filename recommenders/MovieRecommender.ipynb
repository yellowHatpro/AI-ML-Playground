```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
movies_df = pd.read_csv("./dataset/tmdb_5000_movies.csv")
credits_df = pd.read_csv("./dataset/tmdb_5000_credits.csv")
```

```python
credits_df.head()
```

```python
movies_df.head()
```

```python
credits_df.columns = ['id','title','cast','crew']
movies_df = movies_df.merge(credits_df,on='id')
movies_df.head()
```

```python
# Demographic Filtering
C = movies_df['vote_average'].mean()
C
```

```python
M = movies_df['vote_count'].quantile(0.9)
M
```

```python
# Filter out all qualified movies into a new DataFrame
q_movies = movies_df.copy().loc[movies_df['vote_count'] >= M]
q_movies.shape
```

```python
def weighted_rating(x, m=M, c=C):
    v = x['vote_count']
    R = x['vote_average']
    # Calculation based on the IMDB formula
    return (v/(v+m) * R) + (m/(m+v) * c)
```

```python
q_movies['score'] = q_movies.apply(weighted_rating, axis=1)
```

```python
q_movies = q_movies.sort_values('score', ascending=False)
q_movies[['title_x', 'vote_count', 'vote_average', 'score']].head(10)
```

```python
import matplotlib.pyplot as plt
plt.figure (figsize=(12,4))
plt.barh(q_movies['title_x'].head(6),q_movies['score'].head(6), align='center', color='skyblue')
```
